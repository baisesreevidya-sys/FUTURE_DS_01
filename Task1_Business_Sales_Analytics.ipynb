```python
import pandas as pd

df = pd.read_csv('/content/Sample - Superstore.csv', encoding='latin1')

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
# Convert Order Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Create Month-Year column
df['Month'] = df['Order Date'].dt.to_period('M')

# Revenue Trend
revenue_trend = df.groupby('Month')['Sales'].sum()

plt.figure(figsize=(12,5))
revenue_trend.plot(marker='o')
plt.title("Monthly Revenue Trends")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Selling Products
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))
top_products.plot(kind='bar')
plt.title("Top 10 Selling Products")
plt.xlabel("Product Name")
plt.ylabel("Sales")
plt.xticks(rotation=75)
plt.show()
```

```python
# High-value Categories
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color=['green','orange','blue'])
plt.title("High-value Categories")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
# Regional Performance
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
region_sales.plot(kind='bar', color=['purple','red','green','blue'])
plt.title("Regional Performance")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(16, 10))

# Revenue Trends
revenue_trend.plot(ax=ax[0,0], marker='o')
ax[0,0].set_title("Revenue Trends")

# Top-selling Products
top_products.plot(kind='bar', ax=ax[0,1])
ax[0,1].set_title("Top-selling Products")
ax[0,1].tick_params(axis='x', rotation=90)

# High-value Categories
category_sales.plot(kind='bar', ax=ax[1,0])
ax[1,0].set_title("High-value Categories")

# Regional Performance
region_sales.plot(kind='bar', ax=ax[1,1])
ax[1,1].set_title("Regional Performance")

plt.tight_layout()
plt.show()
```

```python
# Monthly Profit Trend
df['Order Date'] = pd.to_datetime(df['Order Date'])
profit_trend = df.groupby(df['Order Date'].dt.to_period('M'))['Profit'].sum()
profit_trend.index = profit_trend.index.astype(str)

plt.figure(figsize=(12,5))
profit_trend.plot(marker='o', color='green')
plt.title("Monthly Profit Trend")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 Customers by Sales
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12,5))
top_customers.plot(kind='bar', color='orange')
plt.title("Top 10 Customers by Sales")
plt.xlabel("Customer")
plt.ylabel("Sales")
plt.xticks(rotation=75)
plt.show()
```

```python
# Sales by Segment
segment_sales = df.groupby('Segment')['Sales'].sum()

plt.figure(figsize=(6,6))
segment_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title("Sales by Segment")
plt.ylabel("")
plt.show()
```

```python
# Sales vs Profit
plt.figure(figsize=(8,5))
plt.scatter(df['Sales'], df['Profit'])
plt.title("Sales vs Profit")
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```
