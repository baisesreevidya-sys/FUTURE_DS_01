```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/bank-additional-full.csv', sep=';')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
contact = df['contact'].value_counts()

print(contact)
```

```python
plt.figure(figsize=(7,5))

contact.plot(kind='bar')

plt.title("Marketing Channel Performance")
plt.xlabel("Contact Type")
plt.ylabel("Number of Customers")

plt.show()
```

```python
campaign = df['poutcome'].value_counts()

print(campaign)
```

```python
plt.figure(figsize=(7,5))

campaign.plot(kind='bar')

plt.title("Previous Campaign Outcome")
plt.xlabel("Campaign Result")
plt.ylabel("Number of Customers")

plt.show()
```

```python
print("Traffic Summary")
print("- Contact Channels Used")
print("- Previous Campaign Performance")
print("- Initial Traffic Analysis Completed")
```

```python
conversion = df['y'].value_counts()

print(conversion)
```

```python
plt.figure(figsize=(6,4))

conversion.plot(kind='bar')

plt.title("Lead to Customer Conversion")
plt.xlabel("Subscription")
plt.ylabel("Number of Customers")

plt.show()
```

```python
conversion_percent = df['y'].value_counts(normalize=True) * 100

print(conversion_percent)
```

```python
plt.figure(figsize=(6,6))

conversion_percent.plot(kind='pie', autopct='%1.1f%%')

plt.title("Lead to Customer Conversion Percentage")
plt.ylabel("")

plt.show()
```

```python
print("Lead to Customer Conversion Summary")
print("- Total Leads Analyzed")
print("- Converted Customers")
print("- Non-Converted Customers")
print("- Conversion Percentage Calculated")
```

```python
dropoff = df['poutcome'].value_counts()

print(dropoff)
```

```python
plt.figure(figsize=(7,5))

dropoff.plot(kind='bar')

plt.title("Conversion Drop-off Analysis")
plt.xlabel("Previous Campaign Outcome")
plt.ylabel("Number of Customers")

plt.show()
```

```python
campaign_contacts = df['campaign'].value_counts().sort_index()

print(campaign_contacts.head(10))
```

```python
plt.figure(figsize=(8,5))

campaign_contacts.head(10).plot(kind='bar')

plt.title("Campaign Contacts Distribution")
plt.xlabel("Number of Contacts")
plt.ylabel("Customers")

plt.show()
```

```python
print("Conversion Drop-off Summary")
print("- Previous Campaign Outcomes Analyzed")
print("- Customer Contact Frequency Reviewed")
print("- Drop-off Stage Identified")
```

```python
channel = df.groupby('contact')['y'].value_counts()

print(channel)
```

```python
plt.figure(figsize=(7,5))

df.groupby('contact')['y'].value_counts().unstack().plot(kind='bar')

plt.title("Channel Performance by Conversion")
plt.xlabel("Contact Channel")
plt.ylabel("Customers")

plt.show()
```

```python
month = df['month'].value_counts()

print(month)
```

```python
plt.figure(figsize=(8,5))

month.plot(kind='bar')

plt.title("Campaign Performance by Month")
plt.xlabel("Month")
plt.ylabel("Customers Contacted")

plt.show()
```

```python
print("Channel Performance Summary")
print("- Contact Channel Performance Analyzed")
print("- Monthly Campaign Performance Reviewed")
print("- Best Performing Marketing Channels Identified")
```
